```python
import requests

url = "https://www.visitmaryland.org/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
}

response = requests.get(url, headers=headers)

print("HTTP Status Code:", response.status_code)
```

The request returned HTTP status code 403 (Forbidden).
The Website is protected by cloudflare and blocks automated Http request, in our case the python libraries make automated requests.
Even adding browser like headers wont work, because the server requires browser cookies to get accepted which the request cannot provide.


```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "lxml")

for element in soup(["script", "style"]):
    element.decompose()

# extract visible text
text = soup.get_text()

# clean text
lines = [line.strip() for line in text.splitlines()]
clean_text = "\n".join(line for line in lines if line)

print(clean_text[:3000])   
```

```python
wiki_url = "https://en.wikipedia.org/wiki/Natural_language_processing"

wiki_response = requests.get(wiki_url, headers=headers)

wiki_soup = BeautifulSoup(wiki_response.text, "lxml")

headings = wiki_soup.find_all(["h1","h2","h3"])

for h in headings:
    print(h.get_text(strip=True))
```

```python
links = wiki_soup.find_all("a")

base = "https://en.wikipedia.org"

for link in links:
    href = link.get("href")
    if href and href.startswith("/wiki/"):
        print(base + href)
```

```python
first_paragraph = wiki_soup.find("p")
paragraph_text = first_paragraph.get_text(strip=True)

print(paragraph_text)
```

```python
with open("nlp_intro.txt", "w", encoding="utf-8") as f:
    f.write(paragraph_text)

print("File saved!")
```
